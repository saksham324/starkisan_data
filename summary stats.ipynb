```python
import pandas as pd
```

```python
means_f = {}
std_f = {}
range_f ={}
means_v = {}
std_v = {}
range_v ={}
for item in list_fruit:
    df = pd.read_excel("fruits/2019/"+item[1:]+"_june_2019.xlsx", )
    df['MODAL_PRICE'] = df['MODAL_PRICE'] / df['WEIGHT_IN_KG']
    df['MIN_PRICE'] = df['MIN_PRICE'] / df['WEIGHT_IN_KG']
    df['MAX_PRICE'] = df['MAX_PRICE'] / df['WEIGHT_IN_KG']
    features = ['MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE']
    means_f[item] = df.groupby('TYPE_OF_GRADE')[features].mean()
    std_f[item] = df.groupby('TYPE_OF_GRADE')[features].std()

    
for item in list_veg:
    df = pd.read_excel("veggies/2019/"+item[1:]+"_june_2019.xlsx")
    df['MODAL_PRICE'] = df['MODAL_PRICE'] / df['WEIGHT_IN_KG']
    df['MIN_PRICE'] = df['MIN_PRICE'] / df['WEIGHT_IN_KG']
    df['MAX_PRICE'] = df['MAX_PRICE'] / df['WEIGHT_IN_KG']
    features = ['MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE']
    means_v[item] = df.groupby('TYPE_OF_GRADE')[features].mean()
    std_v[item] = df.groupby('TYPE_OF_GRADE')[features].std()
```

```python
df_mean_f = pd.DataFrame(columns=['NAME', 'MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE', 'YEAR', 'MONTH'], dtype=float)
df_std_f = pd.DataFrame(columns=['NAME','MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE'])
for key, val in means_f.items():
    dfval = pd.DataFrame(val)
    dfval['NAME'] = key
    dfval['YEAR'] = '2018'
    dfval['MONTH'] = '06'
    df_mean_f = pd.concat([df_mean_f, dfval], axis=0)


for key, val in std_f.items():
    dfval1 = pd.DataFrame(val)
    dfval1['NAME'] = key
    df_std_f = pd.concat([df_std_f, dfval1], axis=0)


df_std_f.columns = [ 'NAME','MIN_PRICE_STD', 'MAX_PRICE_STD', 'MODAL_PRICE_STD']
df_mean_f.columns = ['NAME', 'MIN_PRICE_MEAN', 'MAX_PRICE_MEAN', 'MODAL_PRICE_MEAN', 'YEAR', 'MONTH']

df_fruits_summary = pd.concat([df_mean_f, df_std_f], axis=1)
```
