```python
def getdf(list):
    df = pd.DataFrame(columns=['S', 'NAME', 'ARRIVAL_IN_TONS', 'VARIETY', 'WEIGHT_IN_KG', 'KG', 'PACKING_SIZE', 'TYPE_OF_GRADE', 'MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE', 'DATE'])
    for item in list:
        print(item)
        tables = camelot.read_pdf(item, pages="1-6")
        for i in range(6):
            if item != '2017/23062017.pdf' and item != '2017/24062017.pdf' and item != '2017/27062017.pdf':
                df1 = tables[i].df[10:] if i==0 else tables[i].df[2:]
                df1['DATE'] = item[5:13]
                df1.columns = ['S', 'NAME', 'ARRIVAL_IN_TONS', 'VARIETY', 'WEIGHT_IN_KG', 'KG', 'PACKING_SIZE', 'TYPE_OF_GRADE', 'MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE', 'DATE']
                df = pd.concat([df,df1], axis=0)
            else: 
                df1 = tables[i].df[10:] if i==0 else tables[i].df[2:]
                df1['DATE'] = item[5:13]
                df1.columns = ['S', 'NAME', 'ARRIVAL_IN_TONS', 'VARIETY', 'WEIGHT_IN_KG', 'KG', 'PACKING_SIZE', 'TYPE_OF_GRADE', 'MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE', 'ONE', 'TWO','DATE']
                df1.drop(['ONE', 'TWO'], axis=1)
                df = pd.concat([df,df1], axis=0)
    return df
            
            
```
