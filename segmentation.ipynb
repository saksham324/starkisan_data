```python
import pandas as pd
import sys
import numpy as np
np.set_printoptions(threshold=sys.maxsize)
```

```python
df_fruits = pd.read_excel("fruits_june_2019.xlsx", converters={'S':str})
df_vegetables = pd.read_excel("veggies_june_2019.xlsx")
```

```python
df_fruits
```

```python
array_index = df_fruits['S'].to_numpy()
array_names = df_fruits['NAME'].to_numpy()
array_index_veg = df_vegetables['S'].to_numpy()
array_names_veg = df_vegetables['NAME'].to_numpy()
```

```python
indices = []
indices_veg =[]
names = []
names_veg = []
```

```python
for k in range(len(array_index)):
    if isinstance(array_index[k], str):
        indices.append(array_index[k])
        names.append(array_index[k] + str(array_names[k]))
        
for p in range(len(array_index_veg)):
    if isinstance(array_index_veg[p], str):
        indices_veg.append(array_index_veg[p])
        names_veg.append(array_index_veg[p] + str(array_names_veg[p]))
               
```

```python
names
```

```python
names_veg
```

```python
dff_main = df_fruits.index[df_fruits['S'] + df_fruits['NAME'] == names[1]]
dff_veg = df_vegetables.index[df_vegetables['S'] + df_vegetables['NAME'] == names_veg[1]]
```

```python
main_indices = {}
main_indices_veg = {}
```

```python
for item in names:
    if item not in main_indices:
        main_indices[item] = df_fruits.index[df_fruits['S'] + df_fruits['NAME'] == item]     
        
for veg in names_veg:
    if veg not in main_indices_veg:
        main_indices_veg[veg] = df_vegetables.index[df_vegetables['S'] + df_vegetables['NAME'] == veg]     
```

```python
df_list = {}
df_list_veg = {}
```

```python
for l in range(len(names) - 1):
    if names[l] not in df_list:
        if len(main_indices[names[l]]) == len(main_indices[names[l+1]]):
            for n in range(len(main_indices[names[l]])):
                df_list[names[l]] =[ df_fruits[main_indices[names[l]][n]: main_indices[names[l+1]][n]]]
    else:
        if len(main_indices[names[l]]) == len(main_indices[names[l+1]]):
            for n in range(len(main_indices[names[l]])):
                df_list[names[l]].append(df_fruits[main_indices[names[l]][n] : main_indices[names[l+1]][n]])    

for v in range(len(names_veg) - 1):
    if names_veg[v] not in df_list_veg:
        if len(main_indices_veg[names_veg[v]]) == len(main_indices_veg[names_veg[v+1]]):
            for n in range(len(main_indices_veg[names_veg[v]])):
                df_list_veg[names_veg[v]] =[ df_vegetables[main_indices_veg[names_veg[v]][n]: main_indices_veg[names_veg[v+1]][n]]]
    else:
        if len(main_indices_veg[names_veg[v]]) == len(main_indices_veg[names_veg[v+1]]):
            for n in range(len(main_indices_veg[names_veg[v]])):
                df_list_veg[names_veg[v]].append(df_vegetables[main_indices_veg[names_veg[v]][n] : main_indices_veg[names_veg[v+1]][n]])    
```

```python
list_fruit = []
list_veg = []
for key in df_list:
    df = df_list[key][0]
    for k in range(1, len(main_indices[key])):
        df1 = df_list[key][k]
        df = pd.concat([df, df1], axis=0)
    list_fruit.append(key.replace("/", "_"))
    features = ['MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE']
    df[features].astype('float')
    df.to_excel("fruits/2018/" + key[1:].replace("/", "_") +"_june_2018.xlsx")
    

for veg in df_list_veg:
    df_veg = df_list_veg[veg][0]
    for p in range(1, len(main_indices_veg[veg])):
        print(veg)
        df1 = df_list_veg[veg][p]
        df_veg = pd.concat([df_veg, df1], axis=0) 
    features = ['MIN_PRICE', 'MAX_PRICE', 'MODAL_PRICE']
    df_veg[features].astype('float')
    list_veg.append(veg)
    df_veg.to_excel("veggies/2018/" + veg[1:] +"_june_2018.xlsx")
    
```
