```python
import camelot
import pandas as pd
```

```python
tables = camelot.read_pdf("01012020.pdf", pages="all")
```

```python
df = tables[0].df
df = df[2:]
for i in range(1, 7):
    df1 = tables[i].df
    df1 = df1[2:]
    df = pd.concat([df,df1], axis=0)
```
